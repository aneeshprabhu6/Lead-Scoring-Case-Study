```python
# Drop the specified columns
df_final_with_dummies = df_final_with_dummies.drop(
    ['What is your current occupation_not provided','Lead Origin', 'Lead Source', 'Do Not Email', 'Do Not Call', 
     'Last Activity', 'Country', 'Specialization', 'Specialization_not provided','What is your current occupation', 
     'What matters most to you in choosing a course', 'Search', 'Newspaper Article', 'X Education Forums', 
     'Newspaper', 'Digital Advertisement', 'Through Recommendations','A free copy of Mastering The Interview', 
     'Last Notable Activity'], axis=1)  # Use 'axis=1' to specify columns

# Convert boolean (True/False) values to 1/0 in the entire DataFrame
df_final_with_dummies = df_final_with_dummies.astype(int)

# Display the updated DataFrame
df_final_with_dummies
```

```python
# Import the required library
from sklearn.model_selection import train_test_split
```

```python
#Dropping convetred column
X = df_final_with_dummies.drop(['Converted'], axis=1)  
X.head()
```

```python
# Using iloc to select the 'Converted' column and display the first few rows
y = df_final_with_dummies.iloc[:, df_final_with_dummies.columns.get_loc('Converted')]
print(y.head())
```

```python
# Define the split ratio and random state
split_ratio = 0.7  # 70% for training
random_state_value = 10  # Seed for random state

# Split the dataset using train_test_split with variables
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=split_ratio, test_size=1-split_ratio, random_state=random_state_value)

# Display the shape of the split data
print(f"Training data: X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"Test data: X_test: {X_test.shape}, y_test: {y_test.shape}")
```

```python
# Import MinMax scaler
from sklearn.preprocessing import MinMaxScaler

# Define the numeric columns dynamically
numeric_columns = ['TotalVisits', 'Page Views Per Visit', 'Total Time Spent on Website']

# Create the scaler instance
scaler = MinMaxScaler()

# Use the .transform method on the training set after fitting the scaler to the specified columns
X_train[numeric_columns] = scaler.fit_transform(X_train[numeric_columns])

# Show the transformed data
X_train.head()
```

```python
# To check the correlation among varibles
custom_palette = sns.diverging_palette(230, 20, as_cmap=True)

# Initialize the figure with a different size
plt.figure(figsize=(18, 25))

# Create the heatmap
sns.heatmap(X_train.corr(), annot=True, cmap=custom_palette, fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)

# Set the title for the heatmap
plt.title("Correlation Heatmap for X_train Features", fontsize=20)

# Show the plot
plt.show()
```

```python
# Import 'LogisticRegression'
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()

# Import RFE
from sklearn.feature_selection import RFE

# Running RFE with 15 variables as output
rfe = RFE(estimator=logreg, n_features_to_select=15)
rfe = rfe.fit(X_train, y_train)

# Check the ranking of features
rfe.support_, rfe.ranking_
```

```python
# Features that have been selected by RFE
# Create a DataFrame for better readability
rfe_results = pd.DataFrame({
    'Feature': X_train.columns,
    'Selected': rfe.support_,
    'Ranking': rfe.ranking_
})

# Sort by ranking to view the most important features at the top
rfe_results_sorted = rfe_results.sort_values(by='Ranking', ascending=True)

# Display the sorted results
rfe_results_sorted
```

```python
# Using list comprehension to select the columns where RFE support is True
col = [feature for feature, selected in zip(X_train.columns, rfe.support_) if selected]

# Display the selected columns
col
```

```python
# Selecting columns selected by RFE
X_train = X_train[col]
```

```python
# Importing statsmodels
import statsmodels.api as sm
```

```python
X_train_sm = sm.add_constant(X_train)
logm1 = sm.GLM(y_train, X_train_sm, family = sm.families.Binomial())
res = logm1.fit()
res.summary()
```

```python
# Importing necessary libraries
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a DataFrame to store the VIF values and corresponding features
vif_data = pd.DataFrame()
vif_data['Feature'] = X_train.columns

# Calculate VIF for each feature and store it in the dataframe
vif_data['VIF'] = [variance_inflation_factor(X_train.values, index) for index in range(X_train.shape[1])]

# Round the VIF values to 2 decimal places for clarity
vif_data['VIF'] = vif_data['VIF'].round(2)

# Sort the features by VIF value in descending order to see the most correlated ones first
vif_data_sorted = vif_data.sort_values(by='VIF', ascending=False)

# Display the resulting VIF dataframe
vif_data_sorted
```

```python
# Using zip to combine 'Converted' and 'Conversion_Prob'
y_train_pred_final = pd.DataFrame(list(zip(y_train.values, y_train_pred)), columns=['Converted', 'Conversion_Prob'])

# Display the first 5 rows
y_train_pred_final.head()
```

```python
# Use apply() with a lambda function to classify based on the cutoff of 0.5
y_train_pred_final['Predicted'] = y_train_pred_final['Conversion_Prob'].apply(lambda x: 1 if x > 0.5 else 0)

# Display the first 5 rows
y_train_pred_final.head()
```

```python
# Importing metrics from sklearn for evaluation
from sklearn import metrics

# Compute confusion matrix with swapped arguments
confusion = metrics.confusion_matrix(y_train_pred_final['Predicted'], y_train_pred_final['Converted'])
confusion
```

```python
# Check the overall accuracy
correct_predictions = (y_train_pred_final['Converted'] == y_train_pred_final['Predicted']).sum()
total_predictions = len(y_train_pred_final)
accuracy = correct_predictions / total_predictions
accuracy
```
